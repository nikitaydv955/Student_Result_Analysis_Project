```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("/content/Expanded_data_with_more_features.csv.zip")
dataset.head(5)
```

```python
dataset.describe()
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
# Drop Unnamed Column

dataset = dataset.drop("Unnamed: 0",axis = 1)
print(dataset.head())
```

```python
# change weekly study hour column

dataset["WklyStudyHours"] = dataset["WklyStudyHours"].str.replace("05-Odt","5-10")
dataset.head()
```

```python
# Gender Distribution

plt.figure(figsize = (5,5))
ax = sns.countplot(x = "Gender", data = dataset)
ax.set_title("Gender Distribution")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
#from the above chart we have analysed that:
#the number of female in the data is more than the number of males
```

```python
gb = dataset.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(5,5))
sns.heatmap(gb,annot = True)
plt.title("Relationship between Parent's Education and Student's score")

plt.show()
```

```python
# from the above chart we have concluded that the education of the parent
# have a good impact on their score

```

```python
gb1 = dataset.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb1)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb1,annot = True)
plt.title("Relationship between Parent's Marital Status and Student's score")
plt.show()
```

```python
# from the above chart we have concluded that there is no/negligible impact on the
# student's score due to dear perent's martial status
```

### Distribution of Ethnic Groups

```python
groupA = dataset.loc[(dataset["EthnicGroup"] == "group A")].count()
groupB = dataset.loc[(dataset["EthnicGroup"] == "group B")].count()
groupC = dataset.loc[(dataset["EthnicGroup"] == "group C")].count()
groupD = dataset.loc[(dataset["EthnicGroup"] == "group D")].count()
groupE= dataset.loc[(dataset["EthnicGroup"] == "group E")].count()

l = ["groupA","groupB","groupC","groupD","groupE"]
mlist = [groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
plt.pie(mlist,labels = l, autopct = "%1.2f%%")
plt.title("Distributon of Ethnic Groups")
plt.show()
```
